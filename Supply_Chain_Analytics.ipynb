```
# import packages for analysis and modeling
import pandas as pd  # data frame operations

import numpy as np  # arrays and math functions
from scipy.stats import uniform  # for training-and-test split
import statsmodels.api as sm  # statistical models (including regression)
import statsmodels.formula.api as smf  # R-like model specification
import matplotlib.pyplot as plt  # 2D plotting

import seaborn as sns  # PROVIDES TRELLIS AND SMALL MULTIPLE PLOTTING


# read in Dodgers bobbleheads data and create data frame
sales = pd.read_csv("https://raw.githubusercontent.com/ygeszvain/projects/master/SalesHistory.csv")

# examine the structure of the data frame
print("\nContents of sales data frame ---------------")

# print the first five rows of the data frame
print(pd.DataFrame.head(sales))
salesDF = pd.DataFrame(sales)

salesDF.head()

salesDF12HS = salesDF.loc[salesDF['PNUM'] == "12HS"]
salesDF12LS = salesDF.loc[salesDF['PNUM'] == "12LS"]

salesDFCP = salesDF[['Cost','Price']]

salesDFCP.head()
```

```
#This graph tells us the distribution and the relationship between price and cost
sns.lmplot(x='Cost', y='Price', data=salesDFCP,
           fit_reg=True)
```

```
# Calculate correlations
corr = salesDF.corr()
 
# Heatmap
sns.heatmap(corr)
```

```
# Distribution Plot (12HS)
sns.distplot(salesDF12HS.Cost)
```

```
# Distribution Plot (12HS)
sns.distplot(salesDF12HS.Price)
```

```
# Distribution Plot (12LS)
sns.distplot(salesDF12LS.Cost)
```

```
# Distribution Plot (12LS)
sns.distplot(salesDF12LS.Price)
```

```
#boxplots
sns.boxplot(data=salesDFCP)
```

```
# Density Plot
sns.kdeplot(salesDF12HS.Cost, salesDF12HS.Price)
```

```
# New boxplot using stats_df
sns.boxplot(data=salesDFCP)
```

```
# Joint Distribution Plot
sns.jointplot(x='Cost', y='Price', data=salesDF)
```
